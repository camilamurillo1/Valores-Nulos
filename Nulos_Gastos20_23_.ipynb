```python
#Eliminar primera fila o filas innecesarias
df= pd.read_excel('Gastos y costos 20-23.xlsx', skiprows=[0,1,2,3,4])
df  
```

```python
valores_nulos=df.isnull().sum()
valores_nulos     
```

```python
valores_nulos=df.isnull().sum()
valores_nulos 
```

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en  concreto 
df2["FOLIO"] =df2["FOLIO"].fillna("Sin registro")
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto 
df2["TIPO GASTO"]= df2["TIPO GASTO"].fillna(0)
#Se pone 0 en los faltante de tipo de cambio con la finalidad de no tener valores nulos en el data frame   
df2["TC"]= df2["TC"].fillna(0) 
df2["IMPORTE"]= df2["IMPORTE"].fillna(0) 
df2["IVA"] = df2["IVA"].fillna(df2["IMPORTE"] * 0.16)
df2["TIPO"] =df2["TIPO"].fillna("Sin registro")
df2["POLIZA"] =df2["POLIZA"].fillna("Sin registro")
df2["MP"] =df2["MP"].fillna("Sin registro")
```

```python
valores_nulos=df2.isnull().sum()
valores_nulos 
```
