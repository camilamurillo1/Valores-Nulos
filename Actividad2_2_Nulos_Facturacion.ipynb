##Metodos de sustitucion de valores nulos

```python
df2=df.copy()
```

```python
#SUSTITUIR VALORES NULOS CON PROMEDIO 

df2["CVE_VEND"]=df2["CVE_VEND"].fillna("SC")
df2

#se sustituyeron todos los valores nulos de la columna "CVE_VEND" por la clave "SC"
```

```python
#Comprobamos que ya no hayan valores nulos 

valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#SUSTITUIMOS FECHA_ENT

df2["FECHA_ENT"]=df2["FECHA_ENT"].fillna("00:00:00")
df2

#se sustituyeron todos los valores nulos de la columna "FECHA_ENT" por la hora 00:00:00
```

```python
#Comprobamos 

valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#SUSTITUIMOS FECHA_CANCELA 

df2["FECHA_CANCELA"]=df2["FECHA_CANCELA"].fillna("00:00:00")
df2

#se sustituyeron todos los valores nulos de la columna "FECHA_CANCELA" por la hora 00:00:00
```

```python
#Comprobamos 

valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Comprobamos en todo el data frame

valores_nulos=df2.isnull().sum().sum()
valores_nulos
```
