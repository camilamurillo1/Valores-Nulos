```python
df1=pd.read_csv("inmuebles2.csv", skiprows=1)
df1
```

```python
#axis=1 columnas
#axis=0 filas

df2=df1.drop(["0", "Referencia"], axis=1)
df2

#drop para eliminar 
```

```python
#PARA SUMAR LOS VALORES NULOS POR COLUMNA

valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#para saber los valores nulos en gral de todo el df

valores_nulos=df2.isnull().sum().sum()
valores_nulos
```

##Metodos de sustitucion de valores nulos

```python
df3=df2.copy()
```

```python
#SUSTITUIR VALORES NULOS CON PROMEDIO o media 

df3["Superficie"]=df3["Superficie"].fillna(round(df2["Superficie"].mean(),1)) #el 1 significa cuantos decimales quiero que use 
df3
```

```python
#Comprobamos que ya no hayan valores nulos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SUSTITUIMOS CON LA MEDIANA 

df3["Precio Venta"]=df3["Precio Venta"].fillna(round(df2["Precio Venta"].median(),1))
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SUSTITUIMOS POR UN VALOR NUMERICO EN ESPECIFICO 

df3["Fecha Alta"]=df3["Fecha Alta"].fillna(10)
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#SUSTITUIR VALORES NULOS POR UN STRING 

df3["Fecha Venta"]=df3["Fecha Venta"].fillna("2030-06-19 00:00:00")
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#POR VALORES NO NULOS HACIA ADELANTE "foward fill", es un filtro por columnas, si hay un valor nulo lo rellena con el valor previo

df3["Tipo"]=df3["Tipo"].fillna(method="ffill")
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#Lo mismo que el foward fill pero al revez; backward fill, lo va a rellenar con el que esta despues  

df3["Operación"]=df3["Operación"].fillna(method="bfill")
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["Provincia"]=df3["Provincia"].fillna(method="bfill")
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3["Vendedor"]=df3["Vendedor"].fillna(method="ffill")
df3
```

```python
#Comprobamos 

valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
#Comprobamos en todo el data frame

valores_nulos=df3.isnull().sum().sum()
valores_nulos
```
